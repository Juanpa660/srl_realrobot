```python
import numpy as np
import math
import matplotlib.pyplot as plt
from pid_tuning import Tbot
```

```python
def mse(y_actual, y_predicted):
    return np.round(np.square(np.subtract(y_actual,y_predicted)).mean(), 6)
```

```python
tbot = Tbot()
```

```python
vels =     np.array([0, 0.1, 0.2, 0.1, 0.1, 0.2, 0.2, 0.1,0.05,0.05, 0.1, 0.2, 0.4, 0.4, 0.2, 0])
ang_vels = np.array([0, 0.1, 0.2, 0.2, 0.2, 0.2, 0.4, 0.4, 0.1, 0.1, 0.3, 0.1,-0.1,-0.3, 0.1, 0])
p = 0
d = 0.01 #0.01
ang_p = 0
```

```python
real_vels = []
real_ang_vels = []
signal_vels = []
signal_ang_vels = []
for i in range(len(vels)):
    vel, signal_vel, ang_vel, signal_ang_vel = tbot.step(vels[i], ang_vels[i], p, d, ang_p)
    real_vels.append(vel)
    signal_ang_vels.append(signal_vel)
    real_ang_vels.append(ang_vel)
    signal_ang_vels.append(signal_ang_vel)
```

```python
plt.title(str(mse(vels, real_vels)))
plt.plot(real_vels)
plt.plot(vels)
plt.legend(["Real Speed", "Target Speed"])
plt.show()
```

```python
plt.title(str(mse(ang_vels, real_ang_vels)))
plt.plot(real_ang_vels)
plt.plot(ang_vels)
plt.legend(["Real Speed", "Target Speed"])
plt.show()
```
